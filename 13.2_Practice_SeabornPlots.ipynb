# __SEABORN PLOTS__

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#SEABORN LIBRARY!:
import seaborn as sns

%matplotlib inline
```

```python
diamonds = sns.load_dataset("diamonds")
print(diamonds.shape)
diamonds.head()
```

```python
mpg = sns.load_dataset("mpg")
print(mpg.shape)
mpg.head()
```

__Choose one variable and plot that variable four different ways.__

```python
x = mpg.mpg
sns.distplot(x);
```

```python
sns.catplot(y='mpg', kind='violin', data=mpg)
```

__Choose two continuous variables, and plot them three different ways.__

```python
#sns.relplot(x='weight', y='mpg', data=mpg);

plot1 = sns.lmplot(x='weight',
           y='mpg',
           data=mpg,
           fit_reg=True,
          scatter_kws={'alpha':0.4})
plot1.set_xlabels('Weight (lbs)')
plot1.set_ylabels('mpg')
plt.title("MPG by Weight")
plt.show()
```

```python
plot2 = sns.relplot(x='weight', y='mpg', hue='cylinders', data=mpg, alpha=1)
plot2.set_xlabels('Weight (lbs)')
plot2.set_ylabels('mpg')
plt.title("MPG by Weight")
plt.show()
```

```python
x=mpg.mpg
y=mpg.weight
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k");
```

__Choose one continuous variable and one categorical variable, and plot them six different ways.__

```python
sns.catplot(x='cylinders', y='mpg', kind='violin', data=mpg)
```

```python
sns.catplot(x='cylinders', y='mpg', kind='bar', data=mpg)
```

```python
sns.set(style="whitegrid")

g = sns.pointplot(x="cylinders", y="mpg", data=mpg,
                   height=6, kind="strip", palette="pastel",ci=95,dodge=True,join=False)
plt.title('Pointplot: Price by Cut')
plt.show()
```

```python
sns.catplot(x="clarity", y="price", kind="boxen",
            data=diamonds.sort_values("clarity"));
```

```python
sns.catplot(x='price', y='clarity', kind='violin', data=diamonds)
```

```python
sns.catplot(x="clarity", y="price", kind="bar",
            data=diamonds.sort_values("cut"));
```
